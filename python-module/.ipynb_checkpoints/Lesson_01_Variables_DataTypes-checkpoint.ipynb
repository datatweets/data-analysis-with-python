## Lesson 1 Summary

### What We Learned:

**First Python Code:** `print("Hello World!")`

**Variables:** Boxes for storing data
```python
name = "Ali"
age = 25
```

**Data Types:**
- `int` - Integers (whole numbers)
- `float` - Decimal numbers
- `str` - Strings (text)
- `bool` - True/False

**f-strings:** Easy way to combine text and variables
```python
name = "Ali"
age = 25
print(f"I am {name}, {age} years old")
```

**Type Conversion:**
```python
int("25")      # string → integer
float("3.14")  # string → float
str(25)        # number → string
```

**Getting Input:**
```python
name = input("Your name: ")
age = int(input("Your age: "))
```

### Important Functions:

| Function | Purpose | Example |
|----------|---------|----------|
| `print()` | Display on screen | `print("Hello")` |
| `type()` | Check data type | `type(25)` → `int` |
| `len()` | String length | `len("Hello")` → `5` |
| `int()` | Convert to int | `int("25")` → `25` |
| `float()` | Convert to float | `float("3.14")` |
| `str()` | Convert to string | `str(25)` → `"25"` |
| `input()` | Get user input | `name = input("Name:")` |

### Frequently Asked Questions (FAQ)

**Q: Why does my code give an error?**

A: Most common mistakes:
- Forgetting quotes: `print(Hello)` → `print("Hello")`
- Typo: `prnit` → `print`
- Forgetting type conversion: `int(input("Number: "))`

**Q: What's the difference between `=` and `==`?**

A:
```python
x = 10      # = means "assign" (give value)
x == 10     # == means "compare" (is equal?)
```

**Q: How do I write comments?**

A:
```python
# Single line comment

"""
Multi-line
comment
"""

age = 25    # Comment at end of line
```
